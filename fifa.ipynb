Now that the data is clean we can go ahead and visualize it

```python
sns.set_style('dark')
sns.set_context('notebook')
sns.pairplot(data=df,palette='summer',hue = 'CHAMPION')
```

```python
plt.figure(figsize=(11,5))
plt.title('Years and Goals Scored in World cup')
sns.set_style('dark')
sns.set_context('notebook',font_scale=1)
sns.barplot(data = df ,y='GOALS SCORED',x='YEAR',palette='winter',hue='GOALS SCORED')
plt.show()
```

```python
plt.figure(figsize=(11,5))
plt.title('Years and Mactches Played in World Cup')
sns.set_style('dark')
sns.set_context('notebook',font_scale=1)
sns.barplot(data = df ,y='MATCHES PLAYED',x='YEAR',palette='winter',hue='MATCHES PLAYED')
plt.show()
```

```python
plt.figure(figsize=(11,5))
plt.title('Years and Number of teams that qualified for the world cup')
sns.set_style('dark')
sns.set_context('notebook',font_scale=1)
sns.barplot(data = df ,y='TEAMS',x='YEAR')
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.title('YEARS and AVG GOALS PER GAME at the World cup')
sns.set_style('dark')
sns.set_context('notebook',font_scale=1)
sns.barplot(data = df ,y='AVG GOALS PER GAME',x='YEAR',hue='AVG GOALS PER GAME')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.title('Countries and Number of times they have hosted worldcup')
sns.set_style('darkgrid')
sns.set_context('notebook',font_scale=1)
sns.countplot(data = df ,x='HOST',palette='summer',hue='HOST')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.title('YEARS AND WINNER OF THE WORLD CUP')
sns.set_style('darkgrid')
sns.set_context('notebook',font_scale=1)
sns.countplot(data = df ,x='YEAR',hue='CHAMPION')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.title('HEATMAP FOR NUMERICAL COLUMNS')
sns.set_style('darkgrid')
sns.set_context('notebook',font_scale=1)
heatmap_data = df.select_dtypes(exclude=['object'])
sns.heatmap(heatmap_data.corr(),linewidths=1,linecolor='red',annot=True)
plt.show()
```
